```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import data_table
data_table.enable_dataframe_formatter()
from sklearn.linear_model import LinearRegression
```

```python
data_np = np.loadtxt('data/C1_W2/data/ex1data2.txt',delimiter=',')
(a,b) = np.shape(data_np)
x_train = data_np[:,0:b-1]
y_train = data_np[:,-1].reshape(-1,1)
```

```python
model = LinearRegression();
model.fit(x_train,y_train)
```

```python
y_hat = model.predict(x_train)
```

```python
model.score(x_train,y_train)
```

```python
np.mean(y_train-y_hat)
```

```python
plt.scatter(x_train[:,1],y_hat,marker='.',c='g')
plt.scatter(x_train[:,1],y_train,marker='x',c='r')
plt.show()
```
